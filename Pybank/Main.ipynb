```python
# Script for Python Challenge PyBank  Homework #3
# Instructions


# Create a new GitHub repo called python-challenge. Then, clone it to your computer.
# Inside your local git repository, create a directory for both of the  Python Challenges. Use folder names corresponding to the challenges:
# PyBank and  PyPoll.
# Inside of each folder that you just created, add a new file called main.py. This will be the main script to run for each analysis.
# Push the above changes to GitHub.


# In this challenge, you are tasked with creating a Python script for analyzing the financial records of your company. 
# You will give a set of # financial data called budget_data.csv. 
# The dataset is composed of two columns: Date and Profit/Losses. (Thankfully, your company has rather # lax standards for accounting so the records are simple.)

#Your task is to create a Python script that analyzes the records to calculate each of the following:


# The total number of months included in the dataset
# The total net amount of "Profit/Losses" over the entire period
# The average change in "Profit/Losses" between months over the entire period
# The greatest increase in profits (date and amount) over the entire period
# The greatest decrease in losses (date and amount) over the entire period


# As an example, your analysis should look similar to the one below:


# Financial Analysis
#  ----------------------------
#  Total Months: 86
#  Total: $38382578
#  Average  Change: $-2315.12
#  Greatest Increase in Profits: Feb-2012 ($1926159)
#  Greatest Decrease in Profits: Sep-2013 ($-2196167)

# In addition, your final script should both print the analysis to the terminal and export a text file with the results.

import os
import csv


# Lists to store data
date=[]
revenue=[]
monthly_change=[]


total_months = 0
total_rev = 0
prior_month_rev = 0
min_num = 0
max_num = 0
index = 0


PyBank_csv_path = os.path.join(".", "PyBank_data.csv") 
# .\Desktop\Python_Challenge\PyBank\PyBank_data.csv"

# Open and read csv

with open(PyBank_csv_path,newline='') as PyBank_csvfile:    
    PyBank_file = csv.reader(PyBank_csvfile,delimiter = ',')  
    
    # Read the header row first skip this part if there is no header 
    csv_header = next(PyBank_file)   # this is a function that will store the header information and skip to the next row, next can be nested
    #print(f"Header: {csv_header}")
          
  # Read each row of the data and populate 
    for row in PyBank_file:
        
        #store the first element of file into date as a string
        date.append(row[0])
        
        #store the second element of the file into revenue as an integer
        revenue.append(int(row[1]))


        #Alternate way to calculate total number of months
        total_months = int(total_months) +1
        
        #Calculate the total profit or loss for the entire period by accumulating into variable total_rev
        total_rev = total_rev + int(row[1])
        
        # Store the monthly revenue change as a list
        
        monthly_change.append(int(row[1]) - prior_month_rev)

                
        prior_month_rev = int(row[1])

# Preparing to find the months of greatest and least profitability
# Excluding the first month,  since profitability of the month prior to the first month is assumed zero 
# Therefore not appropriate to include in the the comparison with other months profitability change

monthly_change.pop(0)
date.pop(0)

# print(date,monthly_change)

# Find the dates of greatest and least profitability.  
# If the current minimum number is bigger than the next datapoint, replace the minimum number with the smaller number
# If the current maximum number is smaller than the next datapoint, replace the maximum number with the larger number

for num in monthly_change:
    
    if min_num >= num :
        min_num = num
        min_date = date[index]
    
    if max_num < num :
        max_num = num
        max_date = date[index]
        
    index=index+1       
        

# Calculate the average change as the difference between the last month and first month in the data set, averaged 
# over the total number of months, excluding the first month
# The first month is excluded because we don't have the data for the prior month revenue

ave_monthly_change = round((revenue[-1] - revenue[0])/ (len(date)-1),2)

# Print Summary to terminal
print("Financial Summary")
print("-------------------------------------------------------------")
print("The total number of months is " + str(total_months) + ", from " + str(date[0]) + " to " + str(date[-1]))
#print("The total number of months is " + str(len(date)))
print("The total revenue is $" + str(total_rev))
print("The average monthly revenue = $" + str(ave_monthly_change))
print(f"The month of greatest profitability was ${max_num} on {max_date}")
print(f"The month of least profitability was ${min_num} on {min_date}")

# Output the Summary to a text file


# Open the file in "write" mode ('w') and store the contents in the variable "text" # alias 

with open('./PyBank_output.txt',"w") as text_file:
    print("Financial Summary", file = text_file)
    print("-------------------------------------------------------------", file = text_file)
    print("The total number of months is " + str(total_months) + ", from " + str(date[0]) + " to " + str(date[-1]), file = text_file)

    print("The total revenue is $" + str(total_rev), file = text_file)
    print("The average monthly revenue = $" + str(ave_monthly_change))
    print(f"The month of greatest profitability was ${max_num} on {max_date}", file = text_file)
    print(f"The month of least profitability was ${min_num} on {min_date}", file=text_file)

   
```
